## <center>Data driven prediction models of energy use of appliances in a low-energy house</center>

### Introduction
In this paper various data-driven predictive models for the energy consumption of appliances are discussed. Techniques like data-filtering, removing non-predictive parameters, feature ranking and use of statiscal models are also discussed.
Regression models can give a better understanding of the relationships between different variables and to quantify their impact.
There are two main factors that explain the electricity consumption of a domestic building:
- The type and number of electrical appliances
- The use of the appliances by the occupants

The use of domestic appliance leave a trace of signals like the temperature, humidity, vibrations, light and noise. The  prediction done in this paper was carried out using different data sources and environmental condtions. Data from a nearby airport weather station, temperature and humidity in different rooms in the house from a wireless sensor network and one sub-metered electrical energy consumption have been used to predict the energy use by appliances. Four regression models have been tested.
- Multiple linear regression model
- Support Vector Machine-radial 
- Random forest
- Gradient Boosting Machines



Two Pie charts for Residential electrical energy consumption were displayed. The first chart had percentage of all delivered energy, in the second chart Space heating, Space cooling, Lighting and Water Heating contributions were removed. It was seen that after removing those parameters the energy consumption for other parameters was increased.

The following are few data and methodologies mentioned in the paper to understand appliances energy use:
- Hourly end-use data was collected from 454 houses and 140 commercial buildings in the Pacific Northwest, the end-use loads showed large temporal variation
- Occupancy data from UK households was used and the model uses the Markov Chain Monte Carlo technique to generate synthetic data
- A study used Norton equivalent technique to model residential loads. The modeled loads included appliances such as a refrigerator, PC, laptop, TV and washing machines
- Recent study focused on major household appliances to find daily energy use profiles
- A study focused on the thermal modeling of electrical appliances for highly insulated buildings

<b> Electricity Load Prediction </b>

Models like multiple regression,neural networks, forecasting methods, engineering methods, support vector machines, time series techniques and forecasting methods are used to predict the electricity demand. Parameters such as the time of day, outdoor temperature, month, weekend, holidays, yesterdays consumption, rainfall index etc are used in these models. One study found that being at home correlated with lower appliance efficiency reason being increased use of appliances.
- The energy consumption of appliances represents a significant portion of the aggregated electricity demand of the residential sector (up to 30% see Pie Chart 1), and is also important for powermanagement of the grid.
- The increased number of appliances owned makes it important to identify which ones are the main contributors to the energy consumption.
- The energy consumption of appliances sometimes may include HVAC devices such as air conditioners.
- Weather parameters have proven relevant to predict the electricity consumption in buildings. This consumption includes the HVAC contributions, which may have different trends with respect to temperature. Thus, it is desirable to know whether exterior weather parameters can help in the prediction of appliance energy use.
- For highly insulated buildings, the thermal influence of appliances on internal gains become more important and relevant in building energy performance

#### Results
According to the RFE algorithm, six parameters can significantly reduce the RMSE. These parameter sare: NSM, lights, pressure, RH5, T3 and RH3. The least important predictors are the ones related with week status and day of the week.
In the training set the GBM model had an R2 of 0.97 and the RF an R2 of 0.92. When looking at the ranking for the GBM model with no light information (see Fig. 17), it can be seen that the top predictors are the NSM, Pressure, RH1, RH2, RH3, RH5, T6, RH6, RH4, RH9, T8, T4,and T2. This means that information from the kitchen, living room,laundry room, bathroom, outdoors, office, and bedrooms are the most important

#### Limtations
- The main limitation of this study is that the analysis was done for only one house
- The length of continuous analyzed data
- The problem of optimal location of the wireless indoor sensors was overlooked

#### Conclusions
- Data analysis showed that data from the kitchen, laundryroom, living room and bathrooms had the most important contributions.
- The laundry, kitchen and living rooms are expected to have the highest contributions because of the equipment present.
- Occupancy and occupants activity information could beuseful to improve the prediction and find its relationship with otherparameters
- The GBM models with only weather data ranked pressure as the most important weather variable, followed by the outdoor temperature,dew point temperature, outdoor relative humidity, wind speed and visibility.
- The pressure has a strong prediction power due to its influence on the wind speed and higher rainfall probability which could increase the occupancy of the house.
