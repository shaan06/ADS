## Importing Necessary Packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Reading CSV

```python
energy_dataset = pd.read_csv('E:/ADS/Appliances-energy-prediction-data/energydata_complete.csv')
```

```python
## Describing energy_dataset
energy_dataset.describe().T
```

## Renaming the column names

```python
energy_dataset.rename(columns={'T1':'TempKitchen','RH_1':'HumKitchen','T2':'TempLivingRoom','RH_2':'HumLivingRoom',
                   'T3':'TempLaundryRoom','RH_3':'HumLaundryRoom','T4':'TempOfficeRoom','RH_4':'HumOfficeRoom',
                   'T5':'TempBathRoom','RH_5':'HumBathRoom','T6':'TempOutsideNorth','RH_6':'HumOutsideNorth',
                   'T7':'TempIroningRoom','RH_7':'HumIroningRoom','T8':'TempTeenagerRoom','RH_8':'HumTeenagerRoom',
                   'T9':'TempParentRoom','RH_9':'_HumParentRoom','T_out':'TempOutside','Press_mm_hg':'Pressure',
                   'RH_out':'Humidity'},inplace=True) 
```

```python
##Checking datatypes
energy_dataset.dtypes
```

```python
#Converting date to datetime format
energy_dataset['date'] = pd.to_datetime(energy_dataset['date'])
```

## Checking For Missing Values

```python
energy_dataset.isnull().sum()
```

## Hurray!! There are no missing values

```python
energy_dataset.shape
```

```python
energy_dataset.head(10)
```

```python
## Rounding-off the values
energy_dataset = energy_dataset.round(0)
```

```python
energy_dataset['total_consumption'] = energy_dataset['Appliances'] + energy_dataset['lights']
```

```python
energy_dataset['Appliances'].value_counts().sort_index(ascending=True).plot(figsize = (12,9))
plt.xlabel('Appliance Wh')
plt.ylabel('Frequency')
plt.show()
```

```python
energy_dataset['lights'].value_counts().sort_index(ascending=True).plot(figsize = (12,9))
plt.xlabel('Lights Wh')
plt.ylabel('Frequency')
plt.show()
```

## Plotting the distribution curve of total distribution

```python
print(energy_dataset['total_consumption'].describe().T)
plt.figure(figsize=(9, 8))
sns.distplot(energy_dataset['total_consumption'], color='b', bins=100, hist_kws={'alpha': 0.4});
```

```python
## Adding day
energy_dataset['Day']=energy_dataset['date'].dt.weekday_name
```

```python
## Adding month of the year
energy_dataset['Month'] = energy_dataset['date'].dt.month
```

```python
## Adding time of the day
energy_dataset['hour'] = energy_dataset['date'].dt.hour
```

```python
energy_dataset['NSM'] = energy_dataset['hour']*3600
```

```python
## Plotting Strip plot for the categorical value Day of the Week
g = sns.stripplot(x='Day', y='total_consumption', data=energy_dataset, jitter=True)
plt.xticks(rotation=90)
```

```python
## Plotting Strip plot for the categorical value hour of the day
g = sns.stripplot(x='hour', y='total_consumption', data=energy_dataset, jitter=True)
plt.xticks(rotation=90)
```

```python
## Plotting Strip plot for the categorical value Month of the year
g = sns.stripplot(x='Month', y='total_consumption', data=energy_dataset, jitter=True)
plt.xticks(rotation=90)
```

```python
tdew_point = energy_dataset.groupby('Tdewpoint')['HumKitchen' , 'HumLivingRoom' , 'HumLaundryRoom', 'HumOfficeRoom' , 'HumBathRoom' , 'HumOutsideNorth' , 'HumIroningRoom' , 'HumTeenagerRoom' , '_HumParentRoom'].sum()
tdew_point.plot(kind = 'bar' , figsize = (50,20) , fontsize = 50)
plt.xlabel('TdewPoint', fontsize = 50)
plt.legend(fontsize = 30)
plt.ylabel('Humidity' , fontsize = 50)
plt.show()
```
