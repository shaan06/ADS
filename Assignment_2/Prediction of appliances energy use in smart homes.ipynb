<h1><center>Prediction of applicances energy use in smart homes</center></h1>

This paper presents methods for prediction of energy consumption of different appliances in homes. The energy consumption prediction is done for each appliance in a home,then the forecast will be made for the total energy consumed in a home and, finally, a prediction can be made regarding the households supplied by a certain energy provider. <b> The purpose of this paper is to predict the energy consumption in houses for the next 24h.</b> 

The concept of smart grid appreared as an answer to the new power system challenges.The main purpose of the electrical grid is to assure the energy production necessary for consumers' consumption in a realiable manner. Also, a new concept is developed: <b> demand dispatch</b> : This refers to the ability to control in a precise manner the individual loads at all times, not only at peak period.

<b> Load Management involes two types of load control </b>

- Direct Control: This refers to the classical methods of load control which invole increasing the energy production in case of higher load demand
- Control by cost: The role of cost control is to change the load curve shape in such a way that energy consumption peak decreases, even though the total cosumption for the specific household is the same.

<p><center> The Principal of control by cost </center> </p>

In order to get a better load control, the energy prediction has to go down from total household energy consumption to electrical device consumption. The home management system is very dependent on the forecast of prediction for the services in the house. In this paper, the predictor aims at predicting whether the service will consume energy or not during each hour of the next hour.

For computing the performance of a predictor, test data are first to be considered. In order to evaluate the performance, some concepts have to be defined. Let h be the current hour and e(h) be a binary value which is equal to 0 if the considered appliance is actually consuming energy during the hour h and 1 otherwise.Let Pa(h) be a prediction provided by the predictor a, which is equal to 1 if the considered appliance is predicted as consuming power during the hour h and 0 otherwise.

Any predictor a relies on an historical sliding time window of n hours used to predict the d þ 1 predictions. It can be denoted: a^(h-n->h).The number n has to be adjusted because if it is too large, seasonal phenomena may disappear, and if it is too short, data set will not be sufficient to yield a precise prediction.
